```python
from bs4 import BeautifulSoup

# load the HTML content from a file
with open("HTML.txt", "r", encoding="utf-8") as file:
    soup = BeautifulSoup(file, "lxml")

results = []

# look for all <p> tags
for p in soup.find_all("p"):
    # look for occupation name in <strong> tag
    strong = p.find("strong")
    if not strong:
        continue
    occupation = strong.text.strip()

    ul = p.find_next_sibling("ul")
    if not ul:
        continue

    wage = education = None

    # extract median annual wage and typical entry-level education
    for li in ul.find_all("li"):
        text = li.get_text().strip()
        if "median annual wage" in text.lower():
            wage = text.split("$")[-1].replace(",", "").split()[0]
        if "Typical entry-level education" in text:
            education = li.find("strong")
            if education:
                education = education.text.strip()

    if occupation and wage and education:
        results.append({
            "Occupation": occupation,
            "Median Annual Wage": wage,
            "Typical Entry-Level Education": education
        })

# display the results
for row in results:
    print(row)
```

```python
import csv

with open("occupations_summary.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["Occupation", "Median Annual Wage", "Typical Entry-Level Education"])
    writer.writeheader()
    writer.writerows(results)
```
